```python
import pandas as pd

customers = pd.read_excel("../data/customers.csv.xlsx")
products  = pd.read_excel("../data/products.csv.xlsx")
orders    = pd.read_excel("../data/orders.csv.xlsx")
```

```python
customers['signup_date'] = pd.to_datetime(customers['signup_date'])
orders['order_date'] = pd.to_datetime(orders['order_date'])
```

```python
customers.to_csv("customers_cleaned.csv", index=False)
products.to_csv("products_cleaned.csv", index=False)
orders.to_csv("orders_cleaned.csv", index=False)
```
